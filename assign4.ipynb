```python
curr_x =2
precision = 0.000001
rate=0.01
prv_step_size=1
max_it=1000
it=0
df=lambda x : 2*(x+3)
```

```python
while prv_step_size > precision and it<max_it:
    prv_x=curr_x
    curr_x = curr_x - rate* df(prv_x)
    prv_step_size = abs(curr_x-prv_x)
    it=it+1
    print("itration :", it , "\nx value is :", curr_x)
```
